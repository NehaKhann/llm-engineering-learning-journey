# 📘 Day 6 — Generation Parameters in Large Language Models

In this lesson, we'll explore how different **generation parameters** influence the responses produced by a Large Language Model.

Although the underlying model remains the same, changing parameters such as **temperature**, **max_new_tokens**, and **repetition_penalty** can dramatically alter the model's behavior.

By the end of this notebook, you'll understand how to control the creativity, consistency, and length of generated text.

## 🎯 Learning Objectives

In this notebook you'll learn:

- How GPT-2 generates text
- What **temperature** controls
- Why **sampling** matters
- How **repetition_penalty** improves output quality
- How to compare different generation settings

```python
from transformers import pipeline
import matplotlib.pyplot as plt
```

## 🤖 Load GPT-2

We'll use GPT-2 from Hugging Face to demonstrate how generation parameters influence text generation.

```python
generator = pipeline(
    "text-generation",
    model="gpt2"
)

prompt = "My firewall keeps blocking"
```

## ⚙️ Generation Parameters

We'll compare four different temperature settings.

Each configuration also adjusts the repetition penalty to reduce repeated words and phrases.

```python
parameters = [
    {"name": "Low Temperature (0.3)", "temp": 0.3, "repetition_penalty": 1.10},
    {"name": "Balanced (0.7)", "temp": 0.7, "repetition_penalty": 1.15},
    {"name": "High Temperature (1.2)", "temp": 1.2, "repetition_penalty": 1.20},
    {"name": "Very Creative (1.5)", "temp": 1.5, "repetition_penalty": 1.25},
]
```

## 🚀 Generate Text

We'll generate text using the same prompt while changing the generation parameters.

Observe how the model's writing style changes.

```python
generated_texts = []

print(f'Prompt: "{prompt}"\n')

for param in parameters:

    result = generator(
        prompt,
        max_new_tokens=40,
        temperature=param["temp"],
        do_sample=True,
        repetition_penalty=param["repetition_penalty"],
        no_repeat_ngram_size=2,
        pad_token_id=50256,
    )

    text = result[0]["generated_text"]

    generated_texts.append(text)

    print(param["name"])
    print(text)
    print("-" * 80)
```

## 📊 Visual Comparison

Displaying all generations side-by-side makes it easier to observe how different parameter values influence the model's responses.

```python
plt.figure(figsize=(14,10))

for i, param in enumerate(parameters):

    plt.subplot(2,2,i+1)

    plt.text(
        0.5,
        0.5,
        generated_texts[i],
        ha="center",
        va="center",
        wrap=True,
        fontsize=10
    )

    plt.title(param["name"])
    plt.axis("off")

plt.suptitle("Effect of Generation Parameters on GPT-2")
plt.tight_layout()
plt.show()
```

# 🧠 What Happened?

Notice the differences:

### Low Temperature (0.3)

- More deterministic
- Predictable wording
- Lower creativity

---

### Balanced (0.7)

- Natural responses
- Good balance between creativity and consistency

---

### High Temperature (1.2)

- More diverse wording
- Occasionally unexpected outputs

---

### Very Creative (1.5)

- Highly diverse
- More imaginative
- Can sometimes lose coherence

# 📚 Parameter Summary

| Parameter | Purpose |
|-----------|---------|
| `temperature` | Controls randomness |
| `max_new_tokens` | Controls response length |
| `do_sample` | Enables probabilistic sampling |
| `repetition_penalty` | Discourages repeated words |
| `no_repeat_ngram_size` | Prevents repeated phrases |
